### **Chapter02. 퍼셉트론 (Perception)**

**[단층 퍼셉트론 식]**

$y =\begin{cases}0 & (b + w_{1}x_{1} + w_{2}x_{2} \leq  0)  \\1 & (b + w_{1}x_{1} + w_{2}x_{2} >  0) \end{cases} $

```python
# 가중치 (w : weight) : 각 신호가 결과에 주는 영향력을 제어
# 편향 (b : bias) : 뉴런이 얼마나 쉽게 활성화되는가를 제어
```

```python
## 논리 회로 - AND 게이트 (단층 네트워크)
def AND(x1, x2):
  # 가중치, 임계값(θ) 매개변수
  w1, w2, theta = 0.5, 0.5, 0.7
  tmp = (x1*w1) + (x2*w2)

  if tmp <= theta:
    return 0
  elif tmp > theta:
    return 1
```

```python
## 가중치와 편향 도입
import numpy as np

def AND2(x1, x2):
  x = np.array([x1, x2])      # 입력 신호
  w = np.array([0.5, 0.5])   # 가중치
  b = -0.7                       # 편향
  tmp = np.sum(x*w) + b

  if tmp <= 0:
    return 0
  else:
    return 1
```

```python
## 논리 회로 - NAND 게이트 (단층 네트워크)
def NAND(x1, x2):
  x = np.array(x1, x2)
  w = np.array([-0.5, -0.5])
  b = 0.7
  tmp = np.sum(x * w) + b

  if tmp <= 0:
    return 0
  else:
    return 1

## 논리 회로 - OR 게이트 (단층 네트워크)
def OR(x1, x2):
  # AND 게이트와는 가중치(w, b)의 값만 다름
  x = np.array([x1, x2])
  w = np.array([0.5, 0.5])
  b = -0.2
  tmp = np.sum(x*w) + b

  if tmp <= 0:
    return 0
  else:
    return 1
```

```python
## 논리 회로 - XOR 게이트 (다층 네트워크) / (배타적 논리합 : 두 입력 신호 중 한 쪽이 1일 때만 뉴런 활성화)
def XOR(x1, x2):
  s1 = NAND(x1, x2)
  s2 = OR(x1, x2)
  y = AND(s1, s2)
  return y
```
